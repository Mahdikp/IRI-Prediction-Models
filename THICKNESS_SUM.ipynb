```python
import numpy as np
import pandas as pd

o = pd.read_csv('T2.csv')

df = pd.DataFrame(o,columns = ['STATE_CODE'
                                 ,"SHRP_ID",'CONSTRUCTION_NO',"LAYER_NO",'DESCRIPTION','DESCRIPTION_EXP','LAYER_TYPE'
                                 ,'REPR_THICKNESS',"TOTAL_THICKNESS","AC LAYER_THICKNESS"])
#df = pd.DataFrame(o,columns = ['STATE_CODE',
                               #  "SHRP_ID",'CONSTRUCTION_NO',"LAYER_NO",'DESCRIPTION','DESCRIPTION_EXP','LAYER_TYPE','LAYER_TYPE_EXP'
                              #   ,'REPR_THICKNESS','MATL_CODE_EXP',"TOTAL_THICKNESS","AC LAYER_THICKNESS", 'OVERLAY_THICKNESS'])
```

```python
frames = []
for i in df["STATE_CODE"].unique():
 for j in df["SHRP_ID"].unique():
  for k in df["CONSTRUCTION_NO"].unique():
    T1 = df[(df["STATE_CODE"]==i) & (df["SHRP_ID"]==j) & (df["CONSTRUCTION_NO"]==k)]
    T1["AC LAYER_THICKNESS"].fillna(round(T1['REPR_THICKNESS'].sum(),3), inplace=True)
    frames.append(T1)
result = pd.concat(frames)
```
