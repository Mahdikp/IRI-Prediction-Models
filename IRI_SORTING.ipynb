```python
import numpy as np
import pandas as pd

o = pd.read_csv('IRI.csv')

df = pd.DataFrame(o,columns = ['CONSTRUCTION_NO',"VISIT_DATE","YEAR", "MONTH", "DAY","VISIT_NO",'STATE_CODE','STATE_CODE_EXP'
                                 ,"SHRP_ID","MRI","IRIR","IRI"])
```

```python
df.head()
```

```python
df.tail()
```

```python
frames = []
for i in df["VISIT_NO"].unique():
    T1 = df[(df["VISIT_NO"]==i)]
    T1["IRI"].fillna(round(T1['MRI'].mean(),9), inplace=True)
    frames.append(T1)
result = pd.concat(frames)
```

```python
result
```

```python
df1=result.drop_duplicates(['IRI'])

```

```python
df1.to_csv('IRI_SORTING.csv')
```
